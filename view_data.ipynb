# view data counting-cell

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import cv2 as cv
import pandas as pd
import os
import xml.etree.ElementTree as ET
```

```python
path = "data_count_cell/train/annotations/imagemParaTreinamento1_png.rf.ec63faa7be41590e425a569005406a93.xml"
tree = ET.parse(path)
root = tree.getroot()
img_path = root.find("path").text
name = list()
toa_do = list()
for child in root.iter('object'):
    name.append(child.find("name").text)
    x_min = child.find("bndbox").find("xmin").text
    x_max = child.find("bndbox").find("xmax").text
    y_min = child.find("bndbox").find("ymin").text
    y_max = child.find("bndbox").find("ymax").text
    vitri = [int(x_min),int(x_max),int(y_min),int(y_max)]
    toa_do.append(vitri)
print(f'img_path: {img_path}')
print(f'name: {name}')
print(f'toa_do: {toa_do}')
```

```python
root_img_path = "data_count_cell/train/images/"

img = plt.imread(root_img_path+img_path)
fig, ax = plt.subplots()
ax.imshow(img)
for vitri in toa_do:
    x_min = vitri[0]
    x_max = vitri[1]
    y_min = vitri[2]
    y_max = vitri[3]
    width = x_max-x_min
    height = y_max-y_min
    ax.add_patch(patches.Rectangle((x_min,y_min),width,height,color='r',fill=False))
# plt.imshow()
```
